```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/wyniki_zaglodzenie_n2.csv', header=None)

plt.figure(figsize=(8, 6))
plt.boxplot([data[0], data[1]], tick_labels=['Filozof 1', 'Filozof 2'])
plt.title('Średnie czasy oczekiwania dla 2 filozofów w rozwiązaniu z możliwościa zagłodzenia')
plt.xlabel('Filozof')
plt.ylabel('Średni czas oczekiwania[ms]')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
